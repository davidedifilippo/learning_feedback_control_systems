```python
!pip install control

# importo le librerie
import control
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Funzione di trasferimento ad anello aperto

num = [1000]
den = [1, 2]

F1 = control.tf(num, den)

num = [1]
den = [1, 3]

F2 = control.tf(num, den)

num = [1]
den = [1, 5]

F3 = control.tf(num, den)

F = control.series(F1,F2,F3)

W = control.feedback(F, 1)
print("W1=", W)
```

```python
p,z= control.pzmap(W, plot = True, grid=True)

print(p)

plt.show()
```
