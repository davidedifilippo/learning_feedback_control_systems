```python
# importo le librerie

import control
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Funzione di trasferimento ad anello aperto

# F(s) = 1/s(s^2+3S+2)

num = [1]
den = [1, 3, 2, 0]

F = control.tf(num, den)

print("F=", F)
```

```python
control.pzmap(F, plot = True, grid=True)
```

```python
# Funzione di trasferimento ad anello chiuso

W = control.feedback(F, 1)

print("W=", W)
```

```python
#Trova le radici della W(s) al variare del guadagno K del ramo diretto

control.root_locus(F, initial_gain=0)

plt.show()
```

```python

W = control.feedback(F, 1)
control.pzmap(W, grid=True)

plt.show()
```

```python
# Step Response Feedback System (Tracking System)

t, y = control.step_response(W)

plt.figure(1)
plt.plot(t,y)
plt.title("Step Response Feedback System W(s)")
plt.grid()
```

```python
W = control.feedback(6*F, 1)
control.pzmap(W, grid=True)

plt.show()
```

```python
# Step Response Feedback System (Tracking System)

t, y = control.step_response(W)

plt.figure(1)
plt.plot(t,y)
plt.title("Step Response Feedback System W(s)")
plt.grid()
```

```python
W = control.feedback(10*F, 1)
control.pzmap(W, grid=True)

plt.show()
```

```python
# Step Response Feedback System (Tracking System)

t, y = control.step_response(W)
plt.figure(1)
plt.plot(t,y)
plt.title("Step Response Feedback System W(s)")
plt.grid()
```

```python
# Bode Diagram with Stability Margins

# Frequencies
w_start = 0.01
w_stop = 10
step = 0.01
N = int ((w_stop-w_start )/step) + 1
w = np.linspace (w_start , w_stop , N)

# Bode Plot

w, mag, phase = control.bode(F, w, margins= True)

plt.show()

```

```python
# Calculating stability margins and crossover frequencies

gm , pm , w180 , wc = control.margin(F)

# Convert gm to Decibel

gmdb = 20 * np.log10(gm)
print("wc =", f'{wc:.2f}', "rad/s")
print("w180 =", f'{w180:.2f}', "rad/s")
print("GM =", f'{gm:.2f}')
print("GM =", f'{gmdb:.2f}', "dB")
print("PM =", f'{pm:.2f}', "deg")

# Find when Sysem is Marginally Stable (Kritical Gain - Kc)
Kp = 1
Kc = Kp*gm
print("Kc =", f'{Kc:.2f}')
```
