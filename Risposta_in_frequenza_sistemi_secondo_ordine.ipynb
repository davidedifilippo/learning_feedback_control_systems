```python
# importo le librerie

import control
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Frequenze

w_start = 1
w_stop = 100
step = 0.1

N = int ((w_stop-w_start )/step) + 1
w = np.linspace (w_start , w_stop , N)

#Funzione di trasferimento del secondo ordine


z = 0.707  # smorzamento
w_n = 10  # pulsazione naturale 


num = [1]
den = [1/(w_n*w_n), 2*(1/w_n)*z, 1 ]

F = control.tf(num, den)

control.bode(F, w, dB= True)
plt.show()
```

```python
z = 0.5  # smorzamento

den = [1/(w_n*w_n), 2*(1/w_n)*z, 1 ]

F = control.tf(num, den)
control.bode(F, w, dB= True)
plt.show()
```
