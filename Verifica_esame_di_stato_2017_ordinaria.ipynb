```python
# importo le librerie

import control
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Funzione di trasferimento ad anello aperto

# F(s) = 1/s(s^2+3S+2)

k = 20

num1 = [19, 100]
den1 = [1, 1]

G1 = control.tf(num1, den1)

num2 = [k]
den2 = [1, 10]

G2 = control.tf(num2, den2)

num3 = 0.5
den3= [1,10]

G3 = control.tf(num3, den3)

FH = control.series(G1, G2, G3)



print("FH=", FH)
```

```python
control.pzmap(FH, plot = True, grid=True)
```

```python
#Trova le radici della W(s) al variare del guadagno K del ramo diretto

control.root_locus(FH, initial_gain=0)

plt.show()
```

```python
# Bode Diagram with Stability Margins

# Frequencies
w_start = 0.01
w_stop = 1000
step = 0.01
N = int ((w_stop-w_start )/step) + 1
w = np.linspace (w_start , w_stop , N)

# Bode Plot

w, mag, phase = control.bode(FH, w, margins= True)

plt.show()

```

```python
# Calculating stability margins and crossover frequencies

gm , pm , w180 , wc = control.margin(FH)

# Convert gm to Decibel

gmdb = 20 * np.log10(gm)
print("wc =", f'{wc:.2f}', "rad/s")
print("w180 =", f'{w180:.2f}', "rad/s")
print("GM =", f'{gm:.2f}')
print("GM =", f'{gmdb:.2f}', "dB")
print("PM =", f'{pm:.2f}', "deg")

# Find when Sysem is Marginally Stable (Kritical Gain - Kc)
Kp = 1
Kc = Kp*gm
print("Kc =", f'{Kc:.2f}')
```
